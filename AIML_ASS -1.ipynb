```python
import matplotlib.pyplot as plt

people = ['kiran','arun','vijay','varun']
ages=["25,30,35,40"]
height=[145,151,165,173]
weight=[45,55,65,75]

plt.scatter(weight, height)
plt.title("relationship between weight and height :")
plt.ylabel("eight in meters(100cm)")
plt.xlabel("weight in Kgs")
plt.show()

plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')
plt.bar(x=people,height=weight, width=0.5,
        color='green',edgecolor='k',alpha=0.5)
plt.xlabel("People",fontsize=15)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()

import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='blue',edgecolor='k', alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=14)
plt.xlabel("Weight in kgs",fontsize=14)
plt.show()
```
